This ability to learn from data allows machines to perform a wide range of tasks, from recognizing speech and images to making predictions and decisions in complex environments.

The foundation of machine learning lies in its diverse set of algorithms, which can be broadly categorized into supervised learning, unsupervised learning, and reinforcement learning. Reinforcement learning is a type of machine learning where an agent learns to make decisions by taking actions in an environment to achieve some goals; it's particularly used in areas such as robotics, gaming, and navigation. Each of these paradigms addresses different types of problems but together forms the basis of machine learning, enabling machines to tackle complex tasks with increasing accuracy and sophistication.

By applying supervised learning techniques, such as regression models, data scientists can train algorithms on historical data sets that include features like make, model, year, mileage, and condition, along with their corresponding sale prices. ### 2.2 Log Transformation

**Log Transformation of Target Variable (Price)**:
Log Transformation of the Target Variable, specifically the price of used cars, plays a pivotal role in optimizing the performance of machine learning models in this project. The decision to apply a log transformation stems from a deep understanding of the distribution of car prices within the dataset.

One prominent characteristic observed in the distribution of car prices is its tendency to exhibit a right-skewed pattern. This skewness occurs when a few high-priced cars disproportionately influence the overall distribution, causing it to extend further to the right and deviate from a normal distribution.

To mitigate the adverse effects of this skewness on model performance, we employ a logarithmic transformation on the prices. Typically, the natural logarithm is used for this purpose. By taking the logarithm of the prices, we effectively compress the values of higher-priced cars while spreading out the lower values. This transformation leads to a more symmetrical distribution, aligning closely with the assumptions often made by machine learning models regarding the normal distribution of target variables.

The advantages of log transformation are manifold. Firstly, it enhances the performance of machine learning models by aligning the target variable with the assumptions of normality, thereby facilitating more accurate predictions. Additionally, the transformation helps in mitigating the influence of extreme outliers, particularly those associated with excessively high-priced cars, which could otherwise skew the model's predictions. Moreover, the use of logarithmic transformation enables the interpretation of model errors through metrics such as the Root Mean Squared Logarithmic Error (RMSLE), offering insights into the accuracy of predictions across the entire price range.

In essence, the log transformation of the target variable stands as a crucial preprocessing step, offering substantial benefits in terms of model performance, robustness against outliers, and interpretability of prediction errors, ultimately contributing to the success of the project in predicting used car prices accurately.

**Benefits**: The log transformation offers several benefits:
- Improves model performance as machine learning models often assume normally distributed target variables.
- Mitigates the impact of extreme high-priced outliers.
- Facilitates better interpretation of model errors using metrics like Root Mean Squared Logarithmic Error (RMSLE).

### 2.3 Key Concepts

In machine learning, several key concepts are integral to understanding and effectively applying techniques for used car price prediction. These concepts serve as the foundation for various aspects of model development, including data preprocessing, model selection, and evaluation. Here, we delve deeper into three crucial concepts: Learning Rate, n_estimators, and R² Score.

**Learning Rate**:
The learning rate stands as a pivotal hyperparameter in gradient-based optimization algorithms utilized for training machine learning models. Essentially, it dictates the magnitude of parameter updates during the iterative optimization process. A lower learning rate facilitates smaller updates, leading to slower convergence but reducing the likelihood of overshooting the optimal solution. Conversely, a higher learning rate accelerates convergence but risks overshooting the optimal values, potentially hindering model performance. Striking the right balance in choosing the learning rate is critical for achieving optimal model training and convergence.

**n_estimators**:
The term "n_estimators" holds significance in the domain of ensemble learning methods, particularly in algorithms like Random Forest and AdaBoost. It denotes the number of base learners or decision trees comprising the ensemble model. Increasing the number of estimators often enhances model performance by capturing more intricate patterns present in the data. However, a higher number of estimators also escalates computational complexity, necessitating a trade-off between model performance and computational resources. Thus, determining the optimal value for n_estimators requires careful consideration to achieve the desired balance between model accuracy and computational efficiency.

**R² Score (Coefficient of Determination)**:
The R² score, also referred to as the coefficient of determination, serves as a fundamental metric for evaluating the performance of regression models, including those used in predicting used car prices. It quantifies the proportion of variance in the dependent variable (i.e., car price) that can be explained by the independent variables (e.g., features like mileage, age, etc.). The R² score ranges from 0% to 100%, with higher values indicating a better fit of the model to the data. Essentially, the R² score provides insights into how well the regression model captures the variability in the target variable, facilitating informed decisions regarding model selection and refinement.

In summary, a thorough understanding of key machine learning concepts such as learning rate, n_estimators, and R² score is indispensable for effectively navigating the complexities of used car price prediction. These concepts guide practitioners in making informed decisions at various stages of the modeling process, ultimately contributing to the development of accurate and robust predictive models.

### 3.3 Estimating better iterative imputer method

```python
check_imputer=df2.copy()
check_imputer=check_imputer.drop(columns=['id'])

#object of LabelEncoder
encoder=preprocessing.LabelEncoder()

#function to label the categorical columns ignoring null values 
def encode(data):
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the data for encoding
    impute_reshape = nonulls.reshape(-1,1)
    #encode date
    impute_ordinal = encoder.fit_transform(impute_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data

#loop to label each column or categorical values into numberic
for i in tqdm(range(len(cat_cols))):
    encode(check_imputer[cat_cols[i]])
```

```python
check_imputer.head()
```

Metrics which measure the distance between the model
and the data, like metrics. mean_squared_error ,
are available as neg_mean_squared_error
which return the negated value of the metric.


```python
# Estimate the score on the entire dataset by filling missing values by mean and median

br_estimator = BayesianRidge() #base estimator
N_SPLITS = 6  #defining the cross-validation splitting strategy

score_simple_imputer = pd.DataFrame() #dataframe to store error of imputer
#computing MSE for mean and median
for strategy in ('mean', 'median'):
    X=check_imputer.iloc[:,:-1]
        #y=check_imputer.iloc[:,-1:].values.ravel()
    y = np.ravel(check_imputer.iloc[:,-1:])
    estimator = make_pipeline(
        SimpleImputer(missing_values=np.nan, strategy=strategy),
        br_estimator
    )
    score_simple_imputer[strategy] = cross_val_score(
        estimator, X, y, scoring='neg_mean_squared_error',
        cv=N_SPLITS
    )
```

```python
# Estimate the score on the entire dataset by filling missing values by 4 different iterative imputer


#estimators for IterativeImputer
estimators = [
    BayesianRidge(),
    DecisionTreeRegressor(max_features='sqrt', random_state=0),
    ExtraTreesRegressor(n_estimators=10, random_state=0),
    KNeighborsRegressor(n_neighbors=15)
]
score_iterative_imputer = pd.DataFrame()
for esti in estimators:
        print("Estimating",esti.__class__.__name__)
        imputer = IterativeImputer(esti)
        df_new=check_imputer.copy()
        for col in check_imputer.columns:
            #col=check_imputer.columns[i]
            impute_data=imputer.fit_transform(df_new[col].values.reshape(-1,1))
            impute_data=impute_data.astype('int64')
            impute_data = pd.DataFrame(np.ravel(impute_data))
            df_new[col]=impute_data
        X=df_new.iloc[:,:-1]
        #y=check_imputer.iloc[:,-1:].values.ravel()
        y = np.ravel(df_new.iloc[:,-1:])
        score_iterative_imputer[esti.__class__.__name__] = cross_val_score(esti,X,y,scoring='neg_mean_squared_error',cv=6)
```

```python
#ploting the graphs of scores each method
scores = pd.concat(
    [score_simple_imputer, score_iterative_imputer],
    keys=['SimpleImputer', 'IterativeImputer'], axis=1
)
fig, ax = plt.subplots(figsize=(12, 6))
means = -scores.mean()
errors = scores.std()
means.plot.barh(xerr=errors, ax=ax)
ax.set_title('MSE with Different Imputation Methods')
ax.set_xlabel('MSE (smaller value is better)')
ax.set_yticks(np.arange(means.shape[0]))
ax.set_yticklabels(means.index.tolist())
plt.tight_layout(pad=1)
plt.savefig('different-imputations-method',dpi=None)
plt.show()
```

```python
#it will return the range of the variables and the values outside this range will be outliers
def outliers(arr,col):
    x=sorted(arr[col].values.ravel())
    L_25=25/100*(len(x)+1) #L_p where p=25%
    i_p=int(str(L_25).split(".")[0])
    f_p=int(str(L_25).split(".")[1])
    q1=x[i_p]+f_p*(x[i_p+1]-x[i_p])
    
    L_75=75/100*(len(x)+1) #L_p where p=75%
    i_p=int(str(L_75).split(".")[0])
    f_p=int(str(L_75).split(".")[1])
    q3=x[i_p]+f_p*(x[i_p+1]-x[i_p])
    
    #q1,q3=(arr[col].quantile([0.25,0.75]))
    
    IQR=q3-q1
    x1=q1-1.5*IQR
    x2=q3+1.5*IQR
    return (x1,x2)
```

## 6. Model Implementation

### 6.1 Importing required libraries

```python
import time
from tqdm import tqdm
import matplotlib

#libraries for preprocessing
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#libraries for evaluation
from sklearn.metrics import mean_squared_log_error,r2_score,mean_squared_error
from sklearn.model_selection import train_test_split


#libraries for models
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import Ridge
from sklearn.linear_model import LassoCV,RidgeCV
from yellowbrick.regressor import AlphaSelection

from sklearn.linear_model import Lasso

from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

from sklearn.ensemble import AdaBoostRegressor

import xgboost as xgb
```

### 6.2 Data Soucring

```python
#Loading Dataframe

#df=pd.read_csv("vehiclesFinal2.csv")
df=pd.read_csv("vehiclesFinal.csv")
#df=df.drop('Unnamed: 0',axis=1)
df=df.drop('id',axis=1)

"""df=df.drop('lat',axis=1)
df=df.drop('long',axis=1)
df=df.drop('region',axis=1)"""

df2=df.copy()
df.head()
```

```python
#defining numerical and categorical values

num_col=['year','odometer','long','lat']
cat_cols=['region','manufacturer','model','condition','cylinders','fuel','title_status','transmission','drive','size','type','paint_color']
```

### 6.3 Transforming categorical columns values into integer values

```python
le=preprocessing.LabelEncoder()
df[cat_cols]=df[cat_cols].apply(le.fit_transform)
df.head()
```

### 6.4 Scaling Dataset

```python
#scaling numerical data

norm = StandardScaler()
df['price'] = np.log(df['price'])
df['odometer'] = norm.fit_transform(np.array(df['odometer']).reshape(-1,1))
df['year'] = norm.fit_transform(np.array(df['year']).reshape(-1,1))
df['model'] = norm.fit_transform(np.array(df['model']).reshape(-1,1))

#scaling target variable
q1,q3=(df['price'].quantile([0.25,0.75]))
o1=q1-1.5*(q3-q1)
o2=q3+1.5*(q3-q1)
df=df[(df.price>=o1) & (df.price<=o2)]

df['region'] = norm.fit_transform(np.array(df['region']).reshape(-1,1))
df['lat'] = norm.fit_transform(np.array(df['lat']).reshape(-1,1))
df['long'] = norm.fit_transform(np.array(df['long']).reshape(-1,1))

df.head(2)
```

```python
from sklearn.model_selection import train_test_split

def trainingData(df, n):
    X = df.iloc[:, n]
    y = df.iloc[:, -1:].values.ravel()  # Using .ravel() to convert y to 1D array
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, test_size=0.1, random_state=0)
    return X_train, X_test, y_train, y_test

# Example usage:
X_train, X_test, y_train, y_test = trainingData(df, list(range(len(list(df.columns)) - 1)))
```

```python
#some of models will predict neg values so this function will remove that values
def remove_neg(y_test,y_pred):
    ind=[index for index in range(len(y_pred)) if(y_pred[index]>0)]
    y_pred=y_pred[ind]
    y_test=y_test[ind]
    y_pred[y_pred<0]
    return (y_test,y_pred)

#function for evaluation of model
def result(y_test,y_pred):
    r=[]
    r.append(mean_squared_log_error(y_test, y_pred))
    r.append(np.sqrt(r[0]))
    r.append(r2_score(y_test,y_pred))
    r.append(round(r2_score(y_test,y_pred)*100,4))
    return (r)

#dataframe that store the performance of each model
accu=pd.DataFrame(index=['MSLE', 'Root MSLE', 'R2 Score','Accuracy(%)'])    
```

### 6.5 Model Implementation

### 6.5.1 Linear Regression

```python
#fitting model
LR=LinearRegression()
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
```

```python
#calculating error/accuracy
y_test_1,y_pred_1=remove_neg(y_test,y_pred)
r1_lr=result(y_test_1,y_pred_1)
print('Coefficients: \n', LR.coef_)
print("MSLE : {}".format(r1_lr[0]))
print("Root MSLE : {}".format(r1_lr[1]))
print("R2 Score : {} or {}%".format(r1_lr[2],r1_lr[3]))
accu['Linear Regression']=r1_lr
```

```python
#Ploting feature importance graph

coef = pd.Series(LR.coef_, index = X_train.columns)
imp_coef = coef.sort_values()
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Linear Regression Model")
plt.savefig('Linear-Regression-Feature-Importance.jpg')
plt.show()
```

```python
#Visualization of true value and predicted

df_check = pd.DataFrame({'Actual': y_test_1, 'Predicted': y_pred_1})
df_check = df_check.sample(20)
df_check.plot(kind='bar',figsize=(10,5))
plt.grid(which='major', linestyle='-', linewidth='0.1', color='Green')
plt.title('Performance of Linear Regression')
plt.savefig('Linear-Regression-Performance')
plt.show()
```

### 6.5.2 Ridge regression

```python
#predicting value of alpha

alphas = 10**np.linspace(10,-2,400)
model = RidgeCV(alphas=alphas)
visualizer = AlphaSelection(model)
visualizer.fit(X_train,y_train)
visualizer.show()
```

```python
#model object and fitting model
RR=Ridge(alpha=20.336,solver='auto')
RR.fit(X_train,y_train)
y_pred=RR.predict(X_test)
```

```python
#model evaluation

y_test_2,y_pred_2=remove_neg(y_test,y_pred)
r2_ridge=result(y_test_2,y_pred_2)
print("MSLE : {}".format(r2_ridge[0]))
print("Root MSLE : {}".format(r2_ridge[1]))
print("R2 Score : {} or {}%".format(r2_ridge[2],r2_ridge[3]))
accu['Ridge Regression']=r2_ridge
```

```python
#Visualization of Feature Importance

coef = pd.Series(RR.coef_, index = X_train.columns)
imp_coef = coef.sort_values()
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Ridge Model")
plt.savefig('Ridge-Regression-Feature-Importance.jpg')
plt.show()
```

### 6.5.3 Lasso Regression

### 6.5.4 KNN

```python
#estimating MSLE for k=1-9
R_MSLE=[]
for i in range(1,10):
    KNN=KNeighborsRegressor(n_neighbors=i)
    KNN.fit(X_train,y_train)
    y_pred=KNN.predict(X_test)
    error=np.sqrt(mean_squared_log_error(y_test, y_pred))
    R_MSLE.append(error)
    print("K =",i," , Root MSLE =",error)
```

## Conclusion

The culmination of our data-driven inquiry into used car price prediction is encapsulated in the performance evaluation of multiple regression models. The comparative analysis revealed a spectrum of accuracies, signaling the diverse capabilities of the algorithms tested. Linear, Ridge, and Lasso Regression models displayed a consistent performance, with each achieving approximately 62.7% accuracy. A significant leap in accuracy was observed with the KNN model, which achieved 81.5% accuracy, underscoring the benefits of instance-based learning in capturing the nuances of our dataset. However, the ensemble methods truly excelled; the Random Forest Regressor led the forefront with an impressive 91.06% accuracy, followed closely by the Bagging Regressor and AdaBoost Regressor with more than 90% accuracy, and the XGBoost Regressor with 87.78% accuracy. The Random Forest Regressor emerged as the most effective model, likely due to its ability to handle non-linear relationships and interactions between features without the need for explicit feature engineering. The high accuracy of ensemble models can be attributed to their robustness against overfitting and their capacity for aggregating multiple decision trees to improve the predictive quality.

In conclusion, our exploration highlights the critical role of choosing the right model based on the nature of the data and the prediction task at hand. The ensemble methods, particularly the **Random Forest, have proven to be powerful tools in predicting used car prices with high reliability, offering actionable insights for both sellers and buyers in the used car market**. Future work may explore the integration of additional data sources, feature engineering, and hyperparameter tuning to further enhance the predictive performance of these models.
